```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


csv_file_path = 'updated_data_from_delete_pust_znach.csv'

df = pd.read_csv(csv_file_path,  encoding= 'ISO-8859-1')
```

```python

non_numeric_columns = ['Agency', 'City', 'DATE', 'TIME', 'Road Class', 'H&R', 'Locality', 'School ', 'Rumble Strips', 'CN Zone', 'Light', 'Weather', 'Surf Con', 'Road Char', 'Surface', 'Primary Factor', 'Collision Type', 'Unique Id', 'Traffic Control']
df_numeric = df.drop(columns=non_numeric_columns)

# Построение boxplot для каждого числового признака

# Построение отдельного boxplot для каждого числового признака
plt.figure(figsize=(12, 8))
for i, column in enumerate(df_numeric.columns):
    plt.subplot(4, 4, i + 1)  # Создаем сетку для графиков
    sns.boxplot(x=df_numeric[column])
    plt.title(column)  # Добавляем заголовок с названием признака
    plt.xticks(rotation=45)  # Поворачиваем подписи по оси x для лучшей читаемости
plt.tight_layout()  # Улучшаем расположение графиков
plt.show()

```

```python
duplicates = df.duplicated()
print("Количество дубликатов в датасете:", duplicates.sum())

duplicate_rows = df[duplicates]
print("Дубликаты в датасете:\n", duplicate_rows)

df = df.drop_duplicates()                           
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['Longitude'], df['Latitude'], s=5, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Latitude vs Longitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Weather', y='Primary Factor', data=df)
plt.xticks(rotation=45)
plt.title('Primary Factor and Weather')
plt.show()
```
