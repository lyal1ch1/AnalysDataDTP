    + CLOUDY (Облачно): Небо покрыто облаками, но без осадков.
    + RAIN (Дождь): Присутствует дождь.
    + SNOW (Снег): Присутствует снегопад.
    + SLEET/HAIL/FREEZING (Мокрый снег/Град/Замораживание): Мокрый снег, град или замерзающий дождь.
    + BLOWING SAND/SOIL/SNOW (Ветер с песком/почвой/снегом): Ветер, вызывающий перемещение песка, почвы или снега.
    + FOG/SMOKE (Туман/Дым): Присутствует туман или дым, снижающий видимость.
    + BLOWING (Ветер): Сильный ветер.
    + FOG/SMOKE/SMOG (Туман/Дым/Смог): Присутствует туман, дым или смог.
    + SEVERE CROSS WIND (Сильный боковой ветер): Сильный ветер, боковой к дороге, что может создавать опасные условия для вождения.
1. __Surf Con (Состояние покрытия):__ Состояние дорожного покрытия:

    + ICE (Лед): Дорожное покрытие покрыто льдом.
    + WATER (STANDING OR MOVING) (Вода (стоячая или движущаяся)): Присутствует вода на дороге, как стоячая, так и движущаяся (например, лужи, потоки).
    + LOOSE MATERIAL ON ROAD (Свободные материалы на дороге): На дороге находятся свободные материалы, такие как песок, гравий и т. д.
    + MUDDY (Грязно): Дорога покрыта грязью.
    + LOOSE ROAD MATERIAL (Свободные материалы дороги): Присутствуют свободные материалы, отличные от песка или гравия, на дороге, такие как грязь или мусор.

1. ~~__Rd Junction (Джанкшн дороги):__ Тип перекрестка.~~ но не факт
    + 4-Way (4-сторонний): Обычный перекресток, где четыре дороги встречаются под углом 90 градусов.
    + T (Т-образный): Перекресток, где одна дорога пересекает другую, образуя букву "Т".
    + NONE (Отсутствует): Отсутствие перекрестка или разветвления дорог.
    + RAMP (Рампа): Разветвление, обычно используемое для въезда или выезда с дороги на автомагистраль или с автомагистрали на дорогу.
    + INTERCHANGE (Обмен): Обмен, или развязка, сложная сеть разветвлений и перекрестков, обычно связанная с автомагистралями или магистралями.
    + Y (Y-образный): Перекресток, напоминающий букву "Y", где одна дорога разветвляется на две.
    + FIVE POINT OR MORE (Пять и более точек): Перекресток с пятью или более ветвями.
    + Roundabout (Кольцо): Круговое движение, где дороги вращаются вокруг центрального островка.
    + TRAFFIC CIRCLE/ROUNDABOUT (Кольцо/кольцевое движение): Похоже на круговое движение, где дороги вращаются вокруг центрального островка, может также называться   + "траффик-серкл".
    + RAILROAD CROSSINGS (Железнодорожные переезды): Место пересечения дороги с железнодорожными путями.
    + TRAIL CROSSINGS (Пересечения трасс): Место пересечения дороги с пешеходными или велосипедными трассами.
1. __Road Char (Характер дороги):__ Тип дороги, связанная с происшествием.
    + STRAIGHT/LEVEL (Прямая/Ровная): Прямая дорога без значительного наклона.
    + STRAIGHT/GRADE (Прямая/Склон): Прямая дорога с небольшим наклоном или уклоном.
    + CURVE/GRADE (Изгиб/Склон): Дорога с изгибом и наклоном.
    + CURVE/LEVEL (Изгиб/Ровная): Дорога с изгибом, но без значительного наклона.
    + STRAIGHT/HILLCREST (Прямая/Вершина холма): Прямая дорога, проходящая через холм.
    + NON-ROADWAY CRASH (Происшествие вне дороги): Происшествие, которое произошло вне дороги.
    + CURVE/HILLCREST (Изгиб/Вершина холма): Дорога с изгибом, проходящая через холм. 
1. __Surface (Покрытие):__ Тип покрытия дороги(материал).
    + ASP (Асфальт): Дорога с асфальтовым покрытием.
    + CON (Бетон): Дорога с бетонным покрытием.
    + GRAVEL (Гравий): Дорога с покрытием из гравия.
    + OTHER - EXPLAIN IN NARRATIVE (Другое - пояснение в описании): Другие типы поверхностей, которые могут быть описаны подробнее в текстовом описании или в описательной       + части отчета.
    + OTHER (Другое): Другие типы поверхностей, которые не были указаны в предоставленных категориях.
1. __Primary Factor (Основной фактор):__ Основной фактор, связанный с событием.
    + FAILURE TO YIELD RIGHT OF WAY (Нарушение правил уступки дороги): Не соблюдение правила уступки дороги в соответствии с дорожными знаками и правилами дорожного движения.
    + FOLLOWING TOO CLOSELY (Следование слишком близко): Недостаточное расстояние между автомобилями, следующими друг за другом, что может привести к столкновениям при резком торможении.
    + UNSAFE BACKING (Небезопасное движение задним ходом): Ошибка при движении задним ходом, что может привести к столкновению с другими объектами или транспортными средствами.
    + RAN OFF ROAD RIGHT (Выезд с дороги вправо): Автомобиль покинул дорогу с правой стороны.
    + OTHER (DRIVER) (Другое (виновник)): Другие причины, связанные с водителем, которые не вошли в другие категории.
    + ANIMAL/OBJECT IN ROADWAY (Животное/объект на дороге): Наличие животных или предметов на дороге, которые могут вызвать аварию.
    + DISREGARD SIGNAL/REG SIGN (Пренебрежение сигналом/знаком): Не соблюдение сигналов светофора или дорожных знаков.
    + UNSAFE LANE MOVEMENT (Небезопасное перемещение по полосе движения): Некорректное или опасное переключение полосы движения.
    + IMPROPER TURNING (Неправильный поворот): Ошибка при выполнении поворота.
    + SPEED TOO FAST - WEATHER CONDITIONS (Слишком высокая скорость - погодные условия): Повышенная скорость движения в условиях плохой погоды.
    + UNSAFE SPEED (Небезопасная скорость): Движение с неприемлемой или опасной скоростью.
    + DRIVER DISTRACTED (Отвлеченность водителя): Водитель отвлекся от дороги из-за различных причин, таких как использование мобильного телефона, разговоры с пассажирами  и т. д.
    + LEFT OF CENTER (Левая сторона дороги): Транспортное средство двигалось по левой стороне дороги или пересекло центральную линию разделения движения.
    + IMPROPER LANE USAGE (Неправильное использование полосы движения): Неправильное использование полосы движения, например, несоблюдение правила движения по той или  иной полосе.
    + ROADWAY SURFACE CONDITION (Состояние дорожного покрытия): Плохое состояние дорожного покрытия, такое как ямы, неровности, снег, лед и т. д.
    + OVERCORRECTING/OVERSTEERING (Перекорректирование/перегрузка руля): Перекорректирование руля или избыточное поворачивание водителем, что может привести к потере   контроля над автомобилем.
    + IMPROPER PASSING (Неправильный обгон): Неправильное выполнение маневра обгона другого транспортного средства.
    + OTHER (VEHICLE) (Другое (Транспортное средство)): Другие причины, связанные с транспортным средством, которые не вошли в другие категории.
    + SPEED TOO FAST - WEATHER (Слишком высокая скорость - погода): Повышенная скорость движения в связи с погодными условиями.
    + BRAKE FAILURE OR DEFECTIVE (Отказ или дефект тормозов): Отказ или недостаток в работе тормозной системы транспортного средства.
    + PEDESTRIAN ACTION (Действия пешехода): Действия пешехода, которые могут вызвать дорожное происшествие.
    + OTHER (ENVIRONMENTAL) - EXPLAIN IN NARR (Другие (Экологические) - объяснение в описании): Другие экологические факторы, которые могут быть описаны более подробно в  описании инцидента.
    + DRIVER ASLEEP OR FATIGUED (Водитель засыпает или устает): Водитель засыпает за рулем или испытывает усталость во время вождения.
    + VIEW OBSTRUCTED (Вид затруднен): Водитель имеет ограниченный обзор из-за препятствий или других факторов.
    + DRIVER ILLNESS (Заболевание водителя): Заболевание или нехватка здоровья у водителя, влияющие на его способность управлять автомобилем.
    + WRONG WAY ON ONE WAY (Движение в обратную сторону на односторонней дороге): Движение в обратном направлении на дороге с односторонним движением.
    + CELL PHONE USAGE (Использование мобильного телефона): Использование мобильного телефона во время вождения, что может отвлечь водителя и повлиять на его реакции.
    + TIRE FAILURE OR DEFECTIVE (Отказ или дефект шины): Отказ или дефект в работе шины, что может привести к потере контроля над автомобилем.
    + INSECURE/LEAKY LOAD (Неустойчивая/протекающая нагрузка): Недостаточно укрепленный груз или утечка жидкости с груза, что может представлять опасность для    + безопасности движения.
    + ACCELERATOR FAILURE OR DEFECTIVE (Отказ или дефект ускорителя): Отказ или дефект в работе ускорителя, что может привести к потере контроля над автомобилем.
    + HEADLIGHT DEFECTIVE OR NOT ON (Неисправность фар или их выключение): Неисправность фар или их выключение в условиях, когда это необходимо, что может привести к понижению видимости.
    + OBSTRUCTION NOT MARKED (Препятствие не отмечено): Препятствие на дороге, которое не было правильно обозначено или отмечено, что может представлять опасность для водителей.
    + OTHER (ENVIRONMENTAL) (Другие (Экологические)): Другие экологические факторы, которые могут влиять на безопасность дорожного движения.
    + STEERING FAILURE (Отказ или дефект в работе рулевого управления): Отказ или дефект в работе рулевого управления транспортного средства.
    + ENGINE FAILURE OR DEFECTIVE (Отказ или дефект в работе двигателя): Отказ или дефект в работе двигателя транспортного средства, что может привести к потере мощности  и контроля.
    + OVERSIZE/OVERWEIGHT LOAD (Превышение габаритов/вес груза): Груз, который превышает предельные габариты или весовые нормы, что может повлиять на управляемость  транспортного средства.
    + HOLES/RUTS IN SURFACE (Ямы/колеи на поверхности): Наличие ям или колеи на дорожной поверхности, что может привести к потере контроля над автомобилем.
    + TRAFFIC CONTROL INOPERATIVE/MISSING/OBSC (Отсутствие/неработающий/неуказанный дорожный знак или светофор): Отсутствие, неработающий или неуказанный дорожный знак или светофор, что может привести к путанице или неправильному движению.
    + OTHER TELEMATICS IN USE (Другие телематические средства): Использование других телематических средств или систем, которые могут влиять на вождение или безопасность.
    + OTHER LIGHTS DEFECTIVE (Другие неисправности световых приборов): Другие неисправности световых приборов, не указанные в других категориях.
    + TOW HITCH FAILURE (Отказ сцепного устройства): Отказ сцепного устройства, что может привести к потере контроля над прицепом.
    + LANE MARKING OBSCURED (Значения полосы движения не видны): Полосы движения на дороге не видны из-за плохой видимости или иных причин.   
1. __Collision Type (Тип столкновения):__ Тип столкновения, случившегося в результате происшествия.
    + REAR END (Сзади вперед): Один автомобиль столкнулся с другим сзади.
    + RIGHT ANGLE (Прямой угол): Столкновение произошло под прямым углом, как правило, в перекрестке.
    + RAN OFF ROAD (Съехал с дороги): Автомобиль выехал за пределы дороги.
    + BACKING CRASH (Столкновение задней части автомобиля): Столкновение произошло, когда автомобиль двигался задним ходом.
    + SAME DIRECTION SIDESWIPE (Боковое столкновение в том же направлении): Автомобили столкнулись, двигаясь в одном направлении, боковой стороной.
    + LEFT TURN (Левый поворот): Столкновение произошло во время левого поворота.
    + OTHER - EXPLAIN IN NARRATIVE (Другое - пояснение в описании): Другие типы столкновений, которые могут быть описаны более подробно в описательной части отчета.
    + HEAD ON BETWEEN TWO MOTOR VEHICLES (Лобовое столкновение между двумя автомобилями): Два автомобиля столкнулись лоб в лоб.
    + OPPOSITE DIRECTION SIDESWIPE (Боковое столкновение в противоположном направлении): Автомобили столкнулись, двигаясь в противоположных направлениях, боковой стороной.
    + SIDESWIPE (Боковое столкновение): Автомобили столкнулись боковой стороной.
    + HEAD ON (Лобовое столкновение): Автомобили столкнулись лоб в лоб.
    + RIGHT TURN (Правый поворот): Столкновение произошло во время правого поворота.
    + COLLISION WITH DEER (Столкновение с оленем): Автомобиль столкнулся с оленем или другим животным.
    + LEFT/RIGHT TURN (Левый/правый поворот): Столкновение произошло во время поворота влево или вправо.
    + OTHER (Другое): Другие типы столкновений, которые не вошли в другие категории.
    + NON-COLLISION (Без столкновения): Инцидент, который не привел к фактическому столкновению.
    + DEER (Олень): Столкновение произошло с оленем или другим животным.
    + OPPOSING SIDESWIPE (Боковое столкновение с встречным движением): Автомобили столкнулись боковой стороной при движении навстречу друг другу.
    + COLLISION WITH OBJECT IN ROAD (Столкновение с объектом на дороге): Автомобиль столкнулся с объектом, находящимся на дороге.
    + OBJECT IN ROAD (Объект на дороге): Обнаруженный объект на дороге, который мог стать причиной столкновения.
    + REAR TO REAR (Сзади вперед): Столкновение двух автомобилей задними частями друг в друга.
    + COLLISION WITH ANIMAL OTHER (Столкновение с другим животным): Столкновение с животным, отличным от оленя.
1. __Unique Id (Уникальный идентификатор):__ Уникальный идентификатор происшествия.
1. __Traffic Control (Контроль трафика):__ Информация о средствах контроля движения на дороге.
     + NONE (Отсутствует): Отсутствие контроля за движением на данном участке дороги.
     + LANE CONTROL (Контроль полосы движения): Указания и маркировка на дороге для управления движением в определенных полосах.
     + TRAFFIC CONTROL SIGNAL (Светофор): Светофорный светофор для регулирования потока транспортных средств.
     + STOP SIGN (Стоп-знак): Знак, требующий от водителей остановиться перед пересечением.
     + SIGNAL (Сигнал): Общее обозначение для различных сигналов управления трафиком, включая как светофоры, так и другие типы.
     + NO PASSING ZONE (Зона запрета обгона): Участок дороги, на котором обгон запрещен.
     + YIELD SIGN (Уступи дорогу): Знак, требующий от водителей уступить дорогу другим транспортным средствам.
     + OTHER REGULATORY SIGN/MARKING (Другие регулирующие знаки/разметка): Другие знаки и маркировка, используемые для регулирования движения.
     + OTHER (Другое): Другие типы управления трафиком, которые не вошли в другие категории.
     + FLASHING SIGNAL (Мигающий сигнал): Светофор или сигнал, мигающий для обозначения специальных инструкций или предупреждений.

```python

csv_file_path = 'crash-data-monroe-county-2013-2018-7.csv'
df = pd.read_csv(csv_file_path,  encoding= 'ISO-8859-1')
```

```python
# Удаление столбцов с "плохими" признаками
columns_to_drop = ['House', 'Roadway Id', 'Intersect Rd.', 'Interchange', 'Ramp', 'Property Type', 'Feet From', 'Median', 'Rd Junction', 'Dir', 'CN Type' ]
df.drop(columns=columns_to_drop, inplace=True, axis=1)
```

```python
print("Процент от оставшегося датасета", round(len(df.dropna())/len(df), 2))
df = df.dropna()
df.to_csv('data_without_propuski.csv', index=False)
```

```python
def convert_to_24_hour_format(time):
    if not time.strip() or not any(char.isdigit() for char in time):
        return time
    
    if ':' in time:
        hour, minute_period = time.split(':')
        minute = int(minute_period[:2])
        period = minute_period[-2:].strip().upper()
    else:
        hour = time[:-2]
        if not hour.isdigit(): 
            return time
        minute = int(time[-2:])
        period = "AM"

    hour = int(hour)

    if period == 'AM' and hour == 12:
        hour = 0
    elif period == 'PM' and hour != 12:
        hour += 12

    return '{:02d}:{:02d}'.format(hour, minute)




df["TIME"] = df["TIME"].apply(convert_to_24_hour_format)
```

```python
df.to_csv('updated_data_from_delete_pust_znach.csv', index=False)
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        plt.ylabel('target')  # Переносим сюда, чтобы добавлять метки оси y только в крайнем левом столбце
        if i % ncol == 0:
            plt.ylabel('target')  # Добавляем метки оси y только в крайнем левом столбце
```

```python
target = df['Primary Factor']
pairplot(df_numeric, target)
# plt.figure(figsize=(20, 20))  # Задаем размеры 10 дюймов в ширину и 8 дюймов в высоту
plt.show()
```
